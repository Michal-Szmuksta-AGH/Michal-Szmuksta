***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

Poniższy kod przedstawia funkcję $Q$ przeliczającą jednostki z $g$ na $kg$ i z $km/h$ na $m/s$ i podstawiającą wartości pod wzór. Następnie zakodowane zostało wypisanie wyników obliczeń wraz z przeliczeniem na $kcal$ a także polecenia tworzące wykresy w skali liniowej i logarytmicznej na osi $x$, przedstawiające zależność $Q(v)$ dla ciała o masie $3kg$. Aby wykresy były bardziej czytelne, podpisano obie osie i dodano tytuł.

```python
Q = lambda m,v:(m/1000 * (v*1000/3600)**2)/2
Q_sol = Q(2500, 60)
print('Ilość ciepła wydzielonego podczas hamowania dla ciała o masie m = 2500 g i prędkości v = 60 km/h w dżulach:')
print(Q_sol,'J')
print('Ilość ciepła wydzielonego podczas hamowania dla ciała o masie m = 2500 g i prędkości v = 60 km/h w kilokaloriach:')
print(Q_sol/4186.8,'kcal')
x_v = np.linspace(0, 200)
y_Q = Q(3000,x_v)
plt.figure(1)
plt.plot(x_v,y_Q)
plt.xlim(0,200)
plt.title('Ilość wydzielanego ciepła podczas hamowania\n w zależności od prędkości ciała o masie 3kg')
plt.xlabel(xlabel = 'v[km/h]')
plt.ylabel(ylabel = 'Q(v)[J]')
plt.show()
plt.figure(2)
plt.semilogy(x_v,y_Q)
plt.xlim(0,200)
plt.title('Ilość wydzielanego ciepła podczas hamowania\n w zależności od prędkości ciała o masie 3kg')
plt.xlabel(xlabel = 'v[km/h]')
plt.ylabel(ylabel = 'Q(v)[J]')
plt.show()
```

***Zadanie 4.***  
Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
plt.show()
```

***Zadanie 5.***  
Za pomocą funkcji *compare\_plot* rozwiąż graficznie równanie: $f(x) = g(x)$, gdzie:  
$f(x) = x+2$  
$g(x) = x^2 -2sin(x) + 3$

Poniżej przedstawiono rozwiązanie zadania 4 i 5. Aby generowany wykres był jeszcze bardziej czytelny w implementacji funkcji *compare_plot* zawarto inteligentne ustawianie granic obu osi oraz automatyczne tworzenie siatki na wykresie. 

```python
f = lambda x: x + 2
g = lambda x: x**2 - 2*np.sin(x) + 3
x1 = np.linspace(-1,3)
x2 = x1
y1 = f(x1)
y2 = g(x2)
main.compare_plot(x1,y1,x2,y2,'x','y','Rozwiązanie graficzne równania f(x) = g(x)',
                  '$f(x) = x+2$','$g(x) = x^2 -2sin(x) + 3$')
plt.show()
```

Poniżej przedstawiono rozwiązanie zadania 6 i 7. Aby wygenerować spiralę z użyciem zwykłej funkcji *plt.plot* niezbędne było stworzenie równań parametrycznych x1, y1 oraz x2, y2. Dla poprawienia czytelności wygenerowanego wykresu wymuszono kształt kwadratu dla wykresów składowych, a także oddalono je od siebie aby podpisy osi nie nachodziły na sąsiedni wykres.

```python
ar1 = np.linspace(-50,50,10000)
ar2 = np.linspace(-10,10,10000)
a = 0.2
b = 0.2
x1 = a*np.exp(b*ar1)*np.cos(ar1)
y1 = a*np.exp(b*ar1)*np.sin(ar1)
x2 = a*np.exp(b*ar2)*np.cos(ar2)
y2 = a*np.exp(b*ar2)*np.sin(ar2)

main.parallel_plot(x1,y1,x2,y2,'x','y','x','y','Spirale logarytmiczne','|')
plt.show()
```
